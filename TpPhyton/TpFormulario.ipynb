```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt2
import numpy as np

#Grafico 1
#Realizan deporte x sexo

hombres = []
mujeres = []
otro = []

df1 = pd.read_csv("TpNadia.csv")
size = df1.size

for i in range(1, 34):
    sexo = df1.at[i, "Cual es tu sexo"]
    
    if sexo == "Hombre":
        hombres.append(df1.at[i, "Realizas algun deporte?"])
        
    elif sexo == "Mujer":
        mujeres.append(df1.at[i, "Realizas algun deporte?"])
        
    else:
        otro.append(df1.at[i, "Realizas algun deporte?"])

fig, ax = plt.subplots()

valores = ["Sí", "No"]

x = np.arange(len(valores))
width = 0.35

hombresSi = hombres.count("Sí")
hombresNo = hombres.count("No")

hombres = [hombresSi, hombresNo]

rectsHombre = plt.bar(x - width/3, hombres, width, label="Hombres")

mujeresSi = mujeres.count("Sí")
mujeresNo = mujeres.count("No")

mujeres = [mujeresSi, mujeresNo]

rectsMujer = plt.bar(x + width/3, mujeres, width, label="Mujeres")

otroSi = otro.count("Sí")
otroNo = otro.count("No")

otro = [otroSi, otroNo]

rectsOtro = plt.bar(x + width/3 + width/3, otro, width, label="Otros")

plt.bar_label(rectsHombre, padding=2)
plt.bar_label(rectsMujer, padding=2)
plt.bar_label(rectsOtro, padding=2)

ax.set_ylabel('Cantidad de respuestas')
ax.set_title('Realizan deporte según sexo')
ax.set_xticks(x, valores)
ax.legend()

fig.tight_layout()

plt.show()
```
